import tkinter as tk
from math import sqrt

def click(b):
    """
    Maneja el evento de click de los botones de la calculadora.

    Realiza operaciones básicas, funciones especiales (raíz cuadrada, cuadrado)
    o gestiona la entrada (limpiar, números, punto).
    
    Args:
        b (str): El texto del botón presionado.
    """
    actual = entry.get()
    if b == "=":
        try:
            # Evalúa la expresión matemática actual
            resultado = eval(actual)
            entry.delete(0, tk.END)
            entry.insert(tk.END, str(resultado))
        except Exception:
            entry.delete(0, tk.END)
            entry.insert(tk.END, "Error")
    elif b == "C":
        # Borra toda la entrada
        entry.delete(0, tk.END)
    elif b == "√":
        try:
            # Calcula la raíz cuadrada
            valor = float(actual)
            entry.delete(0, tk.END)
            entry.insert(tk.END, str(sqrt(valor)))
        except:
            entry.delete(0, tk.END)
            entry.insert(tk.END, "Error")
    elif b == "x²":
        try:
            # Calcula el cuadrado
            valor = float(actual)
            entry.delete(0, tk.END)
            entry.insert(tk.END, str(valor**2))
        except:
            entry.delete(0, tk.END)
            entry.insert(tk.END, "Error")
    else:
        # Inserta el carácter presionado
        entry.insert(tk.END, b)

import tkinter as tk
from tkinter import ttk, messagebox
import calendar
from datetime import date

# Constantes de configuración
YEAR_OPTIONS = [2025, 2026]
# Modalidades de turno: (Días_Trabajo, Días_Descanso)
MODALITIES = {'14x14': (14, 14), '7x7': (7, 7), '4x4': (4, 4)}

class PlanificadorTurnos(tk.Tk):
    """
    Clase principal para la aplicación Planificador de Turnos.
    Permite visualizar un calendario y marcar días de trabajo/descanso
    según una modalidad de ciclo (ej: 14x14) y un día de inicio de ciclo.
    """
    def __init__(self):
        """Inicializa la ventana, las variables de estado y la interfaz."""
        super().__init__()
        self.title('Planificador de Turnos')
        self.resizable(False, False)
        
        # Variables de control
        self.selected_year = tk.IntVar(value=2025)
        self.selected_month = tk.IntVar(value=1)
        self.selected_modality = tk.StringVar(value='14x14')
        self.cycle_start = None  # Almacena el objeto date del inicio del ciclo (día 1 de trabajo)
        self.selected_day = None # Almacena el objeto date del día seleccionado
        self.start_time = tk.StringVar(value='08:00')
        self.end_time = tk.StringVar(value='20:00')
        
        self.create_widgets()
        self.draw_calendar()

    def create_widgets(self):
        """Inicializa y posiciona todos los widgets de la interfaz."""
        control_frame = ttk.Frame(self, padding=8)
        control_frame.grid(row=0, column=0, sticky='ew')

        ttk.Label(control_frame, text='Año:').grid(row=0, column=0)
        year_cb = ttk.Combobox(control_frame, values=YEAR_OPTIONS, width=6, state='readonly', textvariable=self.selected_year)
        year_cb.grid(row=0, column=1, padx=4)
        year_cb.bind('<<ComboboxSelected>>', lambda e: self.draw_calendar())

        ttk.Label(control_frame, text='Mes:').grid(row=0, column=2)
        months = [calendar.month_name[i] for i in range(1, 13)]
        month_cb = ttk.Combobox(control_frame, values=months, width=10, state='readonly', textvariable=self.selected_month)
        month_cb.grid(row=0, column=3, padx=4)
        month_cb.bind('<<ComboboxSelected>>', lambda e: self.draw_calendar())

        ttk.Button(control_frame, text='◀', command=self.prev_month).grid(row=0, column=4, padx=5)
        ttk.Button(control_frame, text='▶', command=self.next_month).grid(row=0, column=5)

        ttk.Label(control_frame, text='Modalidad:').grid(row=0, column=6, padx=5)
        mod_cb = ttk.Combobox(control_frame, values=list(MODALITIES.keys()), width=6, state='readonly', textvariable=self.selected_modality)
        mod_cb.grid(row=0, column=7, padx=4)
        mod_cb.bind('<<ComboboxSelected>>', lambda e: self.draw_calendar())

        time_frame = ttk.Frame(self, padding=8)
        time_frame.grid(row=1, column=0)

        ttk.Label(time_frame, text='Inicio:').grid(row=0, column=0)
        ttk.Entry(time_frame, textvariable=self.start_time, width=8).grid(row=0, column=1, padx=4)

        ttk.Label(time_frame, text='Término:').grid(row=0, column=2)
        ttk.Entry(time_frame, textvariable=self.end_time, width=8).grid(row=0, column=3, padx=4)

        btn_frame = ttk.Frame(self, padding=8)
        btn_frame.grid(row=2, column=0)

        ttk.Button(btn_frame, text='Fijar inicio de ciclo', command=self.set_cycle_start).grid(row=0, column=0, padx=4)
        ttk.Button(btn_frame, text='Borrar inicio', command=self.clear_cycle_start).grid(row=0, column=1, padx=4)

        legend_frame = ttk.Frame(self, padding=8)
        legend_frame.grid(row=3, column=0)

        tk.Label(legend_frame, text='Trabajo', bg='blue', fg='white', width=10).grid(row=0, column=0, padx=4)
        tk.Label(legend_frame, text='Descanso', bg='white', relief='solid', width=10).grid(row=0, column=1, padx=4)

        self.calendar_frame = ttk.Frame(self, padding=8)
        self.calendar_frame.grid(row=4, column=0)

    def prev_month(self):
        """Retrocede al mes anterior, actualizando el año si es necesario."""
        m, y = self.selected_month.get(), self.selected_year.get()
        if m == 1:
            if y > YEAR_OPTIONS[0]:
                self.selected_year.set(y - 1)
                self.selected_month.set(12)
        else:
            self.selected_month.set(m - 1)
        self.draw_calendar()

    def next_month(self):
        """Avanza al mes siguiente, actualizando el año si es necesario."""
        m, y = self.selected_month.get(), self.selected_year.get()
        if m == 12:
            if y < YEAR_OPTIONS[-1]:
                self.selected_year.set(y + 1)
                self.selected_month.set(1)
        else:
            self.selected_month.set(m + 1)
        self.draw_calendar()

    def draw_calendar(self):
        """Dibuja el calendario mensual, marcando los días según el ciclo de turnos."""
        # Limpia widgets anteriores
        for widget in self.calendar_frame.winfo_children():
            widget.destroy()

        year = self.selected_year.get()
        month = self.selected_month.get()
        header = tk.Label(self.calendar_frame, text=f'{calendar.month_name[month]} {year}', font=('Arial', 14))
        header.grid(row=0, column=0, columnspan=7, pady=5)

        days = ['Lu', 'Ma', 'Mi', 'Ju', 'Vi', 'Sa', 'Do']
        for i, d in enumerate(days):
            tk.Label(self.calendar_frame, text=d).grid(row=1, column=i)

        month_days = calendar.Calendar(firstweekday=0).monthdayscalendar(year, month)
        for r, week in enumerate(month_days, start=2):
            for c, day in enumerate(week):
                if day == 0:
                    tk.Label(self.calendar_frame, text=' ', width=4, height=2).grid(row=r, column=c)
                else:
                    d_date = date(year, month, day)
                    color = 'blue' if self.is_work_day(d_date) else 'white'
                    fg = 'white' if color == 'blue' else 'black'
                    tk.Button(self.calendar_frame, text=str(day), width=4, height=2, bg=color, fg=fg,
                              command=lambda d=d_date: self.select_day(d)).grid(row=r, column=c, padx=2, pady=2)

    def current_modality(self):
        """Retorna la tupla (días_trabajo, días_descanso) de la modalidad seleccionada."""
        return MODALITIES[self.selected_modality.get()]

    def is_work_day(self, d):
        """
        Determina si una fecha dada (d) es un día de trabajo basándose en el inicio de ciclo.
        
        Args:
            d (datetime.date): La fecha a evaluar.
            
        Returns:
            bool: True si es día de trabajo.
        """
        if not self.cycle_start:
            return False
        work, rest = self.current_modality()
        cycle_length = work + rest
        # Calcula la posición del día dentro del ciclo (0 hasta cycle_length - 1)
        diff = (d - self.cycle_start).days % cycle_length
        return diff < work # True si cae en el rango de días de trabajo

    def select_day(self, d):
        """Almacena el día seleccionado por el usuario y muestra una alerta informativa."""
        self.selected_day = d
        messagebox.showinfo('Día seleccionado', f'Día seleccionado: {d.strftime("%d/%m/%Y")}')

    def set_cycle_start(self):
        """Fija el día seleccionado como el inicio oficial del ciclo de turnos."""
        if not self.selected_day:
            messagebox.showwarning('Atención', 'Seleccione un día primero.')
            return
        self.cycle_start = self.selected_day
        messagebox.showinfo('Inicio fijado', f'Inicio de ciclo: {self.cycle_start.strftime("%d/%m/%Y")}')
        self.draw_calendar()

    def clear_cycle_start(self):
        """Borra la configuración del día de inicio de ciclo."""
        self.cycle_start = None
        messagebox.showinfo('Borrado', 'Inicio de ciclo eliminado.')
        self.draw_calendar()

if __name__ == '__main__':
    app = PlanificadorTurnos()
    app.mainloop()

import matplotlib.pyplot as plt

class Region:
    """Clase para representar una región con su nombre y población."""
    def __init__(self, nombre, poblacion):
        """
        Constructor de la clase Region.
        
        Args:
            nombre (str): El nombre de la región.
            poblacion (int/float): El número de habitantes (en millones).
        """
        self.nombre = nombre
        self.poblacion = poblacion

class Pais:
    """
    Clase para representar un país y sus regiones.
    Permite generar un gráfico de barras de población por región.
    """
    def __init__(self, nombre, regiones):
        """
        Constructor de la clase País.
        
        Args:
            nombre (str): Nombre del país.
            regiones (list): Lista de objetos Region.
        """
        self.nombre = nombre
        self.regiones = regiones

    def graficar(self):
        """
        Genera y muestra un gráfico de barras de población por región.
        """
        # Extraer nombres y poblaciones
        nombres = [r.nombre for r in self.regiones]
        poblaciones = [r.poblacion for r in self.regiones]

        plt.figure(figsize=(8,5))
        plt.bar(nombres, poblaciones)
        plt.title(f"Población por región en {self.nombre}")
        plt.xlabel("Regiones")
        plt.ylabel("Población (millones)")
        plt.show()

import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# --- 1. Cargar y preparar los datos (Dataset Iris) ---
iris = load_iris()
X, y = iris.data, iris.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# --- 2. Configuración del Algoritmo Genético (AG) ---
# Espacio de hiperparámetros a optimizar (los "genes")
PARAM_SPACE = {
    'n_neighbors': [3, 5, 7, 9, 11, 13, 15],  # Hiperparámetro K
    'weights': ['uniform', 'distance']       # Hiperparámetro de peso
}
POPULATION_SIZE = 10  # Tamaño de la población (individuos)
GENERATIONS = 5       # Número de ciclos de evolución

def evaluate_fitness(n_neighbors, weights):
    """
    Función de Aptitud (Fitness).
    Evalúa el rendimiento de un individuo midiendo la precisión (accuracy) del modelo K-NN.
    
    Args:
        n_neighbors (int): El valor K.
        weights (str): La estrategia de ponderación.
        
    Returns:
        float: El score de precisión del modelo (el fitness).
    """
    model = KNeighborsClassifier(n_neighbors=n_neighbors, weights=weights)
    model.fit(X_train, y_train)
    predictions = model.predict(X_test)
    return accuracy_score(y_test, predictions)

def initialize_population():
    """Crea una población inicial de combinaciones de hiperparámetros aleatorios."""
    population = []
    neighbors = PARAM_SPACE['n_neighbors']
    weights = PARAM_SPACE['weights']
    
    for _ in range(POPULATION_SIZE):
        n = np.random.choice(neighbors)
        w = np.random.choice(weights)
        population.append({'n_neighbors': n, 'weights': w, 'fitness': 0.0})
    return population

def genetic_algorithm_search():
    """
    Simula el proceso de Algoritmo Genético: Evaluación, Selección, Cruce y Mutación.
    """
    population = initialize_population()
    best_params = None
    best_fitness = -1

    print("--- Inicio del Algoritmo Genético Simulado ---")

    for gen in range(GENERATIONS):
        # 3. Evaluación (Fitness)
        for individual in population:
            individual['fitness'] = evaluate_fitness(individual['n_neighbors'], individual['weights'])

        # Ordenar por fitness (los mejores primero)
        population.sort(key=lambda x: x['fitness'], reverse=True)
        
        current_best = population[0]
        if current_best['fitness'] > best_fitness:
            best_fitness = current_best['fitness']
            best_params = current_best

        print(f"Generación {gen+1}: Mejor Fitness = {best_fitness:.4f}, Parámetros = {best_params['n_neighbors'], best_params['weights']}")
        
        if gen < GENERATIONS - 1:
            # 4. Selección (Tomamos los 5 mejores como padres)
            parents = population[:5]
            new_population = []

            # 5. Cruce (Crossover) y 6. Mutación
            for i in range(POPULATION_SIZE):
                p1 = np.random.choice(parents)
                
                # Cruce: Heredamos n_neighbors del padre
                n_neighbor_new = p1['n_neighbors']
                weight_new = np.random.choice(PARAM_SPACE['weights'])
                
                # Mutación: 20% de probabilidad de cambiar el valor K
                if np.random.rand() < 0.2:
                    n_neighbor_new = np.random.choice(PARAM_SPACE['n_neighbors'])

                new_population.append({'n_neighbors': n_neighbor_new,
                                       'weights': weight_new,
                                       'fitness': 0.0})
            population = new_population

    print("--- Búsqueda Finalizada ---")
    print(f"Mejores parámetros encontrados: K={best_params['n_neighbors']} con weights='{best_params['weights']}'")
    print(f"Máximo Accuracy (Fitness) alcanzado: {best_fitness:.4f}")

if __name__ == '__main__':
    genetic_algorithm_search()
